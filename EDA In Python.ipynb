```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import io
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df1 = pd.read_csv(io.BytesIO(uploaded['world_population.csv']))
```

```python
df1
```

```python
# setting nums like this .00 after all the numbers
pd.set_option('Display.float_format',lambda x: '%.2f' % x)
```

```python
df1
```

```python
df1.info()
```

```python
# this gives the high level overview of the data as count, mean , std , min etc... very helpful
df1.describe()
```

```python
# with this line we can see what is the count of null vals in each columns
df1.isnull().sum()
```

```python
# with this code we can see the number of unique vals in this data set in each columns
df1.nunique()
```

```python
# wiht this line we're sorting and seeing top 10 vals based on 2022 population
df1.sort_values(by='2022 Population',ascending=False).head(10)
```

```python
# is used to calculate the correlation between all numerical columns in the DataFrame, telling you how strongly two variables are related to each other.
df1.corr(numeric_only=True)
```

```python
#The code creates a heatmap to visually display the correlation between all numerical columns, where color intensity shows the strength of the relationship.
# By seeing it, you can quickly identify which variables are most strongly related to one another.
sns.heatmap(df1.corr(numeric_only=True),annot=True)
plt.rcParams['figure.figsize']=(20,7)
plt.show()
```

```python
# This is doing a groupBy based on the continents and showing their mean vals
df2=df1.groupby('Continent').mean(numeric_only=True)
```

```python
# As i dont know the val "Oceania" so im checking all the rows that has 'Oceania' has
df1[df1['Continent'].str.contains('Oceania')]
```

```python
df2.plot()
```

```python
#  This line will swap the rows and columns of your DataFrame, effectively making the index the new columns and the columns the new index.
df3=df2.transpose()
df3
```

```python
# checking all the columns so we can copy paste easily
df2.columns
```

```python
df2=df1.groupby('Continent')[['1970 Population', '1980 Population', '1990 Population',
       '2000 Population', '2010 Population', '2015 Population',
       '2020 Population', '2022 Population' ]].mean(numeric_only=True).sort_values(by='2022 Population',ascending=False)

df2
```

```python
#  This line will swap the rows and columns of your DataFrame, effectively making the index the new columns and the columns the new index.
df3=df2.transpose()
df3
```

```python
df3.plot()
```

```python
# Now we will use Box plot , it helps us to catch the outliers in our dataset , very helpful
df1.boxplot()
```
